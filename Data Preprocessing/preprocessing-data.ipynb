```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
```

#### Import data and check the data

```python
df = pd.read_csv('C:/Users/bhoir/OneDrive/Desktop/pratice_bl/ML_Tasks/Data Preprocessing/data_preprocessing.csv')
display(df)
```

## a. Handling missing data
#### Handling missing values for Age and Salary with mean imputation

```python
imputer = SimpleImputer(strategy='mean')
df[['Age', 'Salary']] = imputer.fit_transform(df[['Age', 'Salary']])
```

```python
display(df)
```

## b. Handling categorical data

```python
label_encoder_country = LabelEncoder()
df['Country'] = label_encoder_country.fit_transform(df['Country'])
display(df)

label_encoder_purchase = LabelEncoder()
df['Purchased'] = label_encoder_purchase.fit_transform(df['Purchased'])
display(df)
```

## c. Split the dataset into training set and test set

```python
X = df[['Country', 'Age', 'Salary']]
y = df['Purchased']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## d. Feature scaling


```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

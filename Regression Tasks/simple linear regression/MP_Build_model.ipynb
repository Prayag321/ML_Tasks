```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import seaborn as sns
import matplotlib.pylab as plt
```

<h2>Load the Train ,Test dataset </h2>

```python
df_train = pd.read_csv('train.csv')
df_test= pd.read_csv('test.csv')
#check the data
df_train.head(),df_test.head()
```

<h2>Preprocess the data</h2>

```python
# Check for missing values in training and testing data
print("Missing values in training data:")
print(df_train.isnull().sum())

print("\nMissing values in test data:")
print(df_test.isnull().sum())
```

```python
# Remove the Nan value
df_train.dropna(inplace=True)
df_test.dropna(inplace=True)
```

<h2>Split the dataset into training and testing</h2>


```python
X_train, X_test, y_train, y_test =df_train[['x']],df_test[['x']],df_train[['y']],df_test[['y']]
```

```python
display(X_train)
display(y_train)
display(X_test)
display(y_test)
```

<h2>Build the model</h2>

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

<h2>Make prediction on test data.</h2>

```python
y_pred = model.predict(X_test)
```

<h2>Evaluating the model</h2>

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

<h2>Visualize the model</h2>

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.plot(X_test, y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```
